#### Q1: What is Estimation Statistics? Explain point estimate and interval estimate.

* Estimation Statistics: It involves using sample data to make inferences or predictions about a population.
* Point Estimate: It's a single value (such as the sample mean) that serves as an estimate of a population parameter.
* Interval Estimate: It's a range of values (confidence interval) within which the true population parameter is likely to fall.

#### Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.

* Hypothesis Testing: It's a statistical method to make decisions about population parameters based on sample data.
* Importance: It helps us determine if an observed effect in data is statistically significant or likely due to chance. It provides a structured framework for making informed decisions based on data.

#### Q5. Write a Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population.

```python
import scipy.stats as stats
def two_sample_t_test(sample1, sample2, alpha):
    t_stat, p_value = stats.ttest_ind(sample1, sample2)
    if p_value < alpha:
        return "Reject Null Hypothesis"
    else:
        return "Fail to Reject Null Hypothesis"
sample1 = [82, 78, 85, 89, 90]
sample2 = [76, 72, 80, 81, 79]
alpha = 0.05
result = two_sample_t_test(sample1, sample2, alpha)
print(result)
```

#### Q6: What is a null and alternative hypothesis? Give some examples.

* Null Hypothesis (H0): It's a statement of no effect or no difference. Examples: The mean weight of product A is equal to the mean weight of product B.
* Alternative Hypothesis (Ha): It's a statement that contradicts the null hypothesis. Examples: The mean weight of product A is not equal to the mean weight of product B.

#### Q7: Write down the steps involved in hypothesis testing.

1. Formulate null (H0) and alternative (Ha) hypotheses.
2. Choose a significance level (alpha).
3. Collect sample data and calculate test statistic.
4. Calculate p-value.
5. Compare p-value with alpha:
    * If p-value < alpha, reject H0.
    * If p-value >= alpha, fail to reject H0.

#### Q8. Define p-value and explain its significance in hypothesis testing.

* p-value: It's the probability of obtaining a test statistic as extreme as or more extreme than the observed, assuming the null hypothesis is true.
* Significance: A small p-value (typically â‰¤ 0.05) suggests that the observed effect is unlikely due to chance, leading to the rejection of the null hypothesis.
