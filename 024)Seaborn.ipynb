# Seaborn

#### Q1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Here are five plots that can be plotted using the Seaborn library along with their uses:
* Bar Plot: Used to visualize categorical data by representing each category with a bar whose height represents the value.
* Histogram: Used to visualize the distribution of a continuous variable by dividing it into bins and showing the frequency of values in each bin.
* Box Plot: Used to display the distribution of a dataset, showing the median, quartiles, and potential outliers.
* Scatter Plot: Used to show the relationship between two continuous variables, each represented by a point on the plot.
* Pair Plot: Used to visualize pairwise relationships in a dataset with multiple variables.

#### Q2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sb
import matplotlib.pyplot as pt
fmri = sb.load_dataset("fmri")
sb.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
pt.title("Line Plot of Signal over Time")
pt.xlabel("Timepoint")
pt.ylabel("Signal")
pt.legend(title="Event")
pt.show()
```

#### Q3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
titanic = sb.load_dataset("titanic")
pt.figure(figsize=(10, 5))
pt.subplot(1, 2, 1)
sb.boxplot(x="pclass", y="age", data=titanic)
pt.title("Box Plot of Age by Pclass")
pt.xlabel("Pclass")
pt.ylabel("Age")
pt.subplot(1, 2, 2)
sb.boxplot(x="pclass", y="fare", data=titanic)
pt.title("Box Plot of Fare by Pclass")
pt.xlabel("Pclass")
pt.ylabel("Fare")
pt.tight_layout()
pt.show()
```

#### Q4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
diamonds = sb.load_dataset("diamonds")
sb.histplot(data=diamonds, x="price", hue="cut", multiple="stack")
pt.title("Histogram of Price by Cut")
pt.xlabel("Price")
pt.ylabel("Frequency")
pt.legend(title="Cut")
pt.show()
```

#### Q5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
iris = sb.load_dataset("iris")
sb.pairplot(data=iris, hue="species")
pt.title("Pair Plot of Iris Dataset")
pt.show()
```

#### Q6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights = sb.load_dataset("flights")
flights_pivot = flights.pivot_table(index="month", columns="year", values="passengers")
sb.heatmap(data=flights_pivot, cmap="coolwarm", annot=True, fmt="d")
pt.title("Heatmap of Flights Data")
pt.xlabel("Year")
pt.ylabel("Month")
pt.show()
```
